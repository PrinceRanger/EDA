```python
# 19. Does Rating Affect The Revenue?

rating_revenue = IMDB[['Title','Rating' , 'Revenue (Millions)']]
rating_revenue
```

```python
plt.figure(figsize=(12,8))

plt.scatter(rating_revenue['Rating'],rating_revenue['Revenue (Millions)'], label = 'Data Points' , color = 'brown')
plt.title("RELATION BETWEEN RATING AND REVENUES")
plt.xlabel("RATINGS")
plt.ylabel("REVENUE (Millions)")
plt.grid(True)
plt.show()
```

```python
# 20. Classify Movies Based on Ratings [Excellent, Good, and Average]

def grades(Rating):
    if Rating >= 8:
        return 'EXCELLENT'
    elif Rating >= 5:
        return 'GOOD'
    else:
        return 'AVERAGE'
    
IMDB['Grade']=IMDB['Rating'].apply(grades)
IMDB
```
