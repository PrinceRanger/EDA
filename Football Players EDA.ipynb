# FOOTBALL PLAYERS
Exploratory Data Analysis (EDA) is a crucial process for football teams in the modern era. It involves the systematic examination of various data sources to gain insights, identify patterns, and make informed decisions that can significantly impact a team's performance and success. EDA is not limited to just on-field statistics but encompasses a wide range of data, including player performance metrics, injury records, scouting reports, and even fan engagement data. In this introduction, we will explore the importance of EDA for football teams and how it can be leveraged for better decision-making and improved performance.

```python
# import library 

import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# players nationality

players['nationality'].value_counts()
```

```python
# there are lot of players, we are taking top 20 countries 


Country_top_20 = players['nationality'].value_counts().head(20)
Country_top_20
```

```python
# BAR PLOT 

Country_top_20.plot(kind = 'bar',color='DarkGreen' , figsize=(15,5))
plt.xticks(rotation = 90)
plt.title("NUMBERS OF PLAYERS FROM COUNTRY")
plt.xlabel("COUNTRIES")
plt.ylabel("NUMBERS OF PLAYERS")
plt.show()
```

```python
# Short name and wages are saved in new variable to get access easily

players_salary = players[['short_name','wage_eur']]
players_salary.head(20)
```

```python
# Making the data higest on the top

players_salary = players_salary.sort_values(by=['wage_eur'],ascending = False ).head(20)
players_salary
```

```python
players_salary.plot(kind = 'bar',color='DarkGreen' , figsize=(15,5))
plt.xticks(rotation = 90)
plt.title("INCOME OF TOP-20 PLAYERS")
plt.xlabel("NAME")
plt.ylabel("INCOME")
plt.show()
```

```python
# Best TOP-20 Shooting players Ratings

player_shoot_rev = players[['short_name' , 'shooting']]
```

```python
plys_st_wrld_scatt = player_shoot_rev.sort_values(by=['shooting'],ascending=False).head(20)
plys_st_wrld_scatt
```

```python
# plotting

plt.figure (figsize =(15,6))
plt.scatter(plys_st_wrld_scatt['short_name'],plys_st_wrld_scatt['shooting'], color = 'DarkGreen')
plt.xticks(rotation = 90)
plt.title("SHOOTING RATINGS OF THE FOOTBALL PLAYERS")
plt.xlabel("NAME OF THE PLAYERS")
plt.ylabel("SHOOTING RATINGS")
plt.show()
```

```python
# Best TOP-20 Defending players Ratings

player_def_rev = players[['short_name' , 'defending']]
```

```python
plys_df_wrld_scatt = player_def_rev.sort_values(by=['defending'],ascending = False).head(20)
plys_df_wrld_scatt
```

```python
plt.figure (figsize =(15,6))
plt.scatter(plys_df_wrld_scatt['short_name'],plys_df_wrld_scatt['defending'],color = 'DarkGreen')
plt.xticks(rotation = 90)
plt.title("DEFENDING RATINGS OF THE FOOTBALL PLAYERS")
plt.xlabel("NAME")
plt.ylabel("DEFENDING RATINGS")
plt.show()
```
