```python
#import required libraries.

import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the given data set

iris_df = pd.read_csv("P:\SEM_3\Manjunath Sir\IRIS.csv")
iris_df
```

```python
#q1-get the information about data set check for missing values and check for duplicates and drop them

# info
iris_df.info()
```

```python
# shape
iris_df.shape
```

```python
# missing value
iris_df.isna().sum()
```

```python
# duplicate
iris_df.duplicated().sum()
```

```python
# duplicates drop
duplicate = iris_df.drop_duplicates(subset = "species")
duplicate
```

```python
#q2-check for the data set is balanced or not.give your conclusions
iris_df.value_counts("species")

# Dataset is balanced , as all rows in "species" are same.
```

```python
#q3-draw the visualization for target column("class")

sns.countplot(x="species" , data = iris_df)
plt.show()

# Species are counted and ploted by using seaborn
```

```python
#q4-develop the relationship between sepal length and sepal width.conclude the results

plt.figure(figsize=(10,5))
sns.scatterplot(x="sepal_length" , y="sepal_width" , data = iris_df , hue = "species")
plt.legend(loc = "upper left")
plt.show()

# Iris-setosa has the smallest sepal length , while Iris-virginica has the largest sepal length
```

```python
#q5-develop the relationship between petal length and petal width.conclude the results

plt.figure(figsize=(10,5))
sns.scatterplot(x="petal_length" , y="petal_width" , data = iris_df , hue = "species")
plt.legend(loc = "upper left")
plt.show()

# Iris-setosa has the smallest petal length , while Iris-virginica has the largest
```

```python
#q6-plot all the columns relationship using pairplot by using the reference of multivariate analysis.conclude the results

sns.pairplot(data = iris_df , height = 4 , hue = "species")
plt.show()


#species iris-Setosa has the smallest of petals widths and lengths. It also has the smallest sepal length but larger sepal widths.
```

```python
#q7-demonstrate the visualization as part of univariate and bivariate analysis using histogram.conclude the results

fig,p = plt.subplots(2,2 , figsize=(16,10))

p[0,0].set_title("sepal length")
p[0,0].hist(iris_df['sepal_length'] , bins=7);

p[0,1].set_title("sepal width")
p[0,1].hist(iris_df["sepal_width"] , bins=6);

p[1,0].set_title("petal length")
p[1,0].hist(iris_df["petal_length"] , bins=5);

p[1,1].set_title("petal width")
p[1,1].hist(iris_df["petal_width"] , bins=5);

plt.show()


# Highest frequency of sepal length is around 34 , which is between 5.5 to 6.0
# Highest frequency of sepal width is around 48 , which is between 3.0 to 3.5
# Highest frequency of petal length is 50 , which is between 1 to 2
# Highest frequency of sepal length is approximatly 50 , which is between 0 to 0.5
```

```python
#q9-how do you handle correlation.

iris_df.corr()
```

```python
#q10-develop the correlation between all numeric variables in dataset

plt.figure(figsize=(10,5))
sns.heatmap(iris_df.corr() , annot=True)
plt.show()
```

```python
#q11-with help of visualization plotting show how categorical values are distributed over numerical values

def graph(y):
    sns.boxplot(x = "species" , y = y , data = iris_df)

    
plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
graph("sepal_length")

plt.subplot(2,2,2)
graph("sepal_width")

plt.subplot(2,2,3)
graph("petal_length")

plt.subplot(2,2,4)
graph("petal_width")

plt.show()
```
