# ANALYSIS 

```python
# MOST RUN MAKER

most_run=ipl_batsman[['Batsman','Run','team']]
most_run
```

```python
most_run_maker_sum = pd.DataFrame(most_run.groupby(['Batsman'])['Run'].sum().sort_values(ascending=False)).head(20)
most_run_maker_sum
```

```python
most_run_maker_sum.plot(kind ='bar', color='#FA8128' , figsize = (15,6))
plt.xticks(rotation=90)
plt.title('MAXIMUM RUN MAKER OF SEASON')
plt.xlabel('NAME OF BATSMAN')
plt.ylabel('NUMBERS OF RUNS')
plt.grid(True)
plt.show()
```

```python
# MOST WICKET TAKER

most_wicket=ipl_bolwer[['Bowler','wicket','team']]
most_wicket
```

```python
most_wicket_taker_sum = pd.DataFrame(most_wicket.groupby(['Bowler'])['wicket'].sum().sort_values(ascending=False)).head(20)
most_wicket_taker_sum
```

```python
most_wicket_taker_sum.plot(kind ='bar', color='purple' , figsize=(15,6))
plt.xticks(rotation=90)
plt.title('MAXIMUM WICKET TAKER OF SEASON')
plt.xlabel('NAME OF BOLWER')
plt.ylabel('NUMBERS OF WICKETS')
plt.grid(True)
plt.show()
```

```python
#Man of the Match

man_of_the_match_IPL = ipl_matches[['match_no','Home_team','Away_team','winner','man_of_the_match']]
man_of_the_match_IPL
```

```python
man_of_the_match_IPL_sum = pd.DataFrame(man_of_the_match_IPL.groupby(['man_of_the_match'])['man_of_the_match'].count().sort_values(ascending=False)).head(20)
man_of_the_match_IPL_sum
```

```python
man_of_the_match_IPL_sum.plot(kind = 'bar',color='Green' , figsize=(15,6))
plt.xticks(rotation=90)
plt.title('TOP-20 MAN OF MATCH TITLE')
plt.xlabel('NAME OF PLAYERS')
plt.ylabel('NUMBERS OF TIMES MAN OF THE MATCH')
plt.grid(True)
plt.show()
```
